# Hadoop MapReduce

Run a Python streaming MapReduce job that counts events by type.

![Architecture](../assets/aws_data_engineering_architecture.svg)

```python
from pathlib import Path
import os, sys, json, subprocess
ROOT=Path('..').resolve()
DRY_RUN=os.getenv('DRY_RUN','true').lower()!='false'
print({'root':str(ROOT),'dry_run':DRY_RUN})
```

Validate mapper and reducer locally.

```python
sample='event_id,customer_id,region,event_type,amount,event_ts\n1,101,South,purchase,10,2026-01-01\n2,102,West,view,0,2026-01-01\n3,103,South,purchase,20,2026-01-01\n'
import subprocess
m=subprocess.run([sys.executable,str(ROOT/'jobs/hadoop/mapper.py')],input=sample,text=True,capture_output=True,check=True)
r=subprocess.run([sys.executable,str(ROOT/'jobs/hadoop/reducer.py')],input=''.join(sorted(m.stdout.splitlines(True))),text=True,capture_output=True,check=True)
print(r.stdout)
```

EMR Hadoop Streaming command.

```python
print('hadoop-streaming -files mapper.py,reducer.py -mapper mapper.py -reducer reducer.py -input s3://<bucket>/data/raw/customer_events.csv -output s3://<bucket>/outputs/mapreduce')
```
